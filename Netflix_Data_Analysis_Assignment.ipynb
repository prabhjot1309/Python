 2. Data Cleaning:

```python
# Check and handle missing values appropriately
df.isnull().sum()
```

```python
(df.isnull().mean() * 100)
```

```python
df.columns
```

```python
df_cleaned = df.dropna() #to remove rows with missing values
df_cleaned = df.dropna(subset=['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'listed_in', 'description']) #Remove rows where specific column having NaN
df_cleaned
```

```python
print(df.isnull().sum())
```

```python
# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['date_added']
```

```python
# Extract year from 'date_added' and store in a new column
df['year_added'] = df['date_added'].dt.year
df['year_added']
```

 3. Data Analysis:

```python
# Count of Movies vs. TV Shows
print("Count of Movies vs TV Shows:",df['type'].value_counts())
```

```python
# Top 5 countries by content count
print("\nTop 5 countries by content count:")
print(df['country'].value_counts().head(5))
```

```python
#- Most common genres listed- Year with the highest content added
all_geners = df['listed_in'].dropna().str.split(',')
genre_series = all_geners.explode()
print("\nTop 10 most common genres:")
print(genre_series.value_counts().head(10))

#Year with highest content added
print("\nYear with highest content added:")
print(df['year_added'].value_counts().idxmax())
```

```python
# Total content added each year
print("\nTotal content added each year:")
print(df['year_added'].value_counts().sort_index())
```

Data Visualization (Using Matplotlib)

```python
# Bar chart: Movies vs. TV Shows
type_counts = df['type'].value_counts()
plt.figure(figsize=(8, 4))
type_counts.plot(kind='bar',color=['skyblue','salmon'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Movies vs. TV Shows')
plt.tight_layout()
plt.show()
```

```python
#- Pie chart: Distribution of top 5 genres
genres = df['listed_in'].dropna().str.split(',').explode().value_counts().head(5)
plt.figure(figsize=(6,6))
genres.plot(kind='pie',startangle=140,colors=plt.cm.Pastel1.colors)
plt.title('Distribution of Top 5 Genres')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
#- Line graph: Content added over the years
content_per_year = df['year_added'].value_counts().sort_index()
plt.figure(figsize=(8, 4))
plt.plot(content_per_year.index, content_per_year.values, marker='o',color='green',linestyle='--')
plt.xlabel('Year')
plt.ylabel('Content Count')
plt.title('Content Added Over The Years')
plt.grid(True,alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#- Horizontal bar chart: Top 5 content-producing countries
top_countries = df['country'].value_counts().head(5)
plt.figure(figsize=(8, 4))
top_countries.plot(kind='barh',color='orange')
plt.xlabel('Content Count')
plt.gca().invert_yaxis()
plt.ylabel('Country')
plt.title('Top 5 Content-Producing Countries')
plt.tight_layout()
plt.show()
```
